## 3. ***Data Wrangling***

Some names of the variables are not relevant and not easy to understand. Some data may have data entry errors, and some variables may need data type conversion and we need to fix this issue in the data and this is reffered as the "Data Wrangling" process.

### Data Wrangling Code

````python
not_no_var = []  # Collection of variable which are not having numerical values. {list(df.select_dtypes(include="object"))}
for column in df:
  if df.dtypes[column] == "object":      #df['Churn'].dtype
    not_no_var.append(column)
print(not_no_var)
````

````python
for var in not_no_var:
  unique = df[var].unique() # Checking unique values of diffrent columns
  print(f"{unique} these are the unique values of {var} column")
````

````python
df['International plan'] = df['International plan'].replace({"Yes":1,"No":0})
df['Voice mail plan'] = df['Voice mail plan'].replace({"Yes":1,"No":0}) # Converting object type into int
````

````python
churn = pd.DataFrame(df['Churn'].value_counts())
churn['Churn Rate'] = round(churn*100/len(df),2) # Check for how many customer leave the company (Churn)
churn
````

````python
for col in df: # column wise churn rate
  churn_col = pd.DataFrame(df.groupby(col)['Churn'].value_counts()).rename(columns = {'Churn':'Count of Churn'})
  churn_col['churn rate'] = round(churn_col*100/len(df),2)
  print(churn_col)
````
